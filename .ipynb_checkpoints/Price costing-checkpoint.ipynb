```python
import csv

def calculate_vcd_cost(width, height):
    #Variables Used
    scrap_add_50 = width + 50
    scrap_add_110 = width + 110

    components = [
        {"name": "Num Louver 1.5", "quantity": height / 150, "unit_price": 0},
        {"name": "louvers Size 1.5", "quantity": ((scrap_add_50 * 175) / 2976800) * (height / 150), "unit_price": 170},
        {"name": "outer 1.5", "quantity": ((scrap_add_110 * 175) / 2976800) * 4, "unit_price": 170},
        {"name": "Corner SHEET 1 MM", "quantity": ((scrap_add_110 * 175) / 2976800) * 2, "unit_price": 110},
        {"name": "BUSH", "quantity": (height / 150) * 2, "unit_price": 3},
        {"name": "BOULT 25 X6", "quantity": (height / 150) * 4, "unit_price": 0.04},
        {"name": "Self Lock Nut", "quantity": (height / 150) * 8, "unit_price": 0.03},
        {"name": "Boult 20x6", "quantity": (height / 150) * 4, "unit_price": 0.05},
        {"name": "Boult 45x8", "quantity": 1, "unit_price": 0.04},
        {"name": "Boult 20x8", "quantity": 1, "unit_price": 0.04},
        {"name": "Washer 8", "quantity": 2, "unit_price": 0.03},
        {"name": "Butrfly Nuy 8", "quantity": 1, "unit_price": 0.03},
        {"name": "SELF DRILL SCREW 7/7.16", "quantity": (scrap_add_50 / 300) * 4, "unit_price": 0.035},
        {"name": "Metal Handle 3", "quantity": 1, "unit_price": 2},
        {"name": "Metal Modulator 1.2", "quantity": 1, "unit_price": 0.5},
        {"name": "Side Links", "quantity": (height / 150) * 2, "unit_price": 1.5},
        {"name": "Spendal Rod 12.7 x65", "quantity": ((height / 150) * 2) - 1, "unit_price": 1.5},
        {"name": "Spendal Rod 12.7 x150", "quantity": 1, "unit_price": 3},
        {"name": "Spendal Rod 12.7 x3000 (BIG SIZES)", "quantity": 0, "unit_price": 7},
        {"name": "Spray Paint", "quantity": 1, "unit_price": 2},
    ]
    
    cost_breakdown = [
        {"component": comp["name"], "quantity": comp["quantity"], "unit_price": comp["unit_price"], "cost": comp["quantity"] * comp["unit_price"]}
        for comp in components
    ]
    total_cost = sum(item["cost"] for item in cost_breakdown)
    return total_cost, cost_breakdown

def calculate_vcd_selling_price(material_cost):
    # Constants for VCD-specific calculation
    participation_rate = 0.35
    direct_labor = material_cost * participation_rate
    overhead = direct_labor * 0.6  # Example calculation
    total_cost = material_cost + direct_labor + overhead

    # Profit margin
    profit_margin = 0.25
    selling_price = (total_cost * profit_margin) + total_cost

    product_code = "VCD-F"
    description = "Manual Volume Control Damper in G.I Material (16 Gauge)"

    return {
        "total_cost": total_cost,
        "selling_price": selling_price,
        "product_code": product_code,
        "description": description,
    }

def calculate_mfsd_cost(width, height):
    #Variables used in formulas
    scrap_add_50 = width + 50
    scrap_add_110 = width + 110
    weight_50 = height + 50
    number_of_motor = 1
    height_divide_150 = height / 150

    components = [
        {"name": "Num Louver", "quantity": height_divide_150, "unit_price": 0},
        {"name": "louvers Size", "quantity": ((scrap_add_50 * 175) / 2976800) * height_divide_150, "unit_price": 185},
        {"name": "outer", "quantity": ((scrap_add_110 * 175) / 2976800) * 4, "unit_price": 185},
        {"name": "Corner SHEET", "quantity": ((scrap_add_110 * 175) / 2976800) * 2, "unit_price": 110},
        {"name": "BUSH", "quantity": height_divide_150 * 2, "unit_price": 3},
        {"name": "BOULT 25 X6", "quantity": height_divide_150 * 4, "unit_price": 0.04},
        {"name": "Self Lock Nut", "quantity": height_divide_150 * 8, "unit_price": 0.03},
        {"name": "Boult 20x6", "quantity": height_divide_150 * 4, "unit_price": 0.05},
        {"name": "SELF DRILL SCREW 7/7.16", "quantity": (scrap_add_50 / 300) * 4, "unit_price": 0.035},
        {"name": "Side Links", "quantity": height_divide_150 * 2, "unit_price": 6},
        {"name": "Spendal Rod 12.7 x65", "quantity": (height_divide_150 * 2) - 1, "unit_price": 1.5},
        {"name": "Num Of Motor", "quantity": 1, "unit_price": 0},
        {"name": "beLIMO Motor", "quantity": number_of_motor * 0, "unit_price": 0},  
        {"name": "BELIMO SENSOR", "quantity": 0, "unit_price": 0},  
        {"name": "beLIMO Holder", "quantity": number_of_motor * 2, "unit_price": 0.5},  
        {"name": "Beliom Boult 75x6", "quantity": number_of_motor * 4, "unit_price": 0.04},  
        {"name": "Plain Nut", "quantity": number_of_motor * 4, "unit_price": 0.02},  
        {"name": "Spray Paint", "quantity": 2, "unit_price": 5},  
        {"name": "Rubber", "quantity": scrap_add_50 * height_divide_150, "unit_price": 0},
        {"name": "JAMSHIL", "quantity": weight_50 * 2, "unit_price": 0},
        {"name": "Rod 3000 x 12.7", "quantity": scrap_add_50, "unit_price": 0},
        {"name": "MFSD sleeve", "quantity": ((width + 60) + (height + 60) * 400) / 2976800, "unit_price": 0},  
    ]
    
    cost_breakdown = [
        {"component": comp["name"], "quantity": comp["quantity"], "unit_price": comp["unit_price"], "cost": comp["quantity"] * comp["unit_price"]}
        for comp in components
    ]
    total_cost = sum(item["cost"] for item in cost_breakdown)
    return total_cost, cost_breakdown

def calculate_fd_cost(width, height):
    # Variables used in formulas
    scrap_add_50 = width + 50
    width_60 = width + 60
    height_60 = height + 60

    components = [
        {"name": "Num Louver", "quantity": height / 50, "unit_price": 0},
        {"name": "louvers Size", "quantity": ((scrap_add_50 * 50) / 2976800) * (height / 50), "unit_price": 185},
        {"name": "outer", "quantity": ((scrap_add_50 * 185) / 2976800) * 4, "unit_price": 185},
        {"name": "Fusibel LINK", "quantity": 1, "unit_price": 0},
        {"name": "Fusibel LINK Huck", "quantity": 2, "unit_price": 0},
        {"name": "ALUMINUM Rivet", "quantity": 4, "unit_price": 0},
        {"name": "Spray Paint", "quantity": 2, "unit_price": 5},
        {"name": "FD sleeve", "quantity": ((width_60 + height_60 * 400) / 2976800), "unit_price": 0},
    ]
    
    cost_breakdown = [
        {"component": comp["name"], "quantity": comp["quantity"], "unit_price": comp["unit_price"], "cost": comp["quantity"] * comp["unit_price"]}
        for comp in components
    ]
    total_cost = sum(item["cost"] for item in cost_breakdown)
    return total_cost, cost_breakdown

def calculate_mvcd_cost(width, height):
    # Variables used in formulas
    scrap_add_50 = width + 50
    scrap_add_110 = width + 110
    weight_50 = height + 50
    number_of_motor = 1
    height_divide_150 = height / 150

    components = [
        {"name": "Num Louver", "quantity": height_divide_150, "unit_price": 0},
        {"name": "louvers Size", "quantity": ((scrap_add_50 * 175) / 2976800) * height_divide_150, "unit_price": 185},
        {"name": "outer", "quantity": ((scrap_add_110 * 175) / 2976800) * 4, "unit_price": 185},
        {"name": "Corner SHEET", "quantity": ((scrap_add_110 * 175) / 2976800) * 2, "unit_price": 110},
        {"name": "BUSH", "quantity": height_divide_150 * 2, "unit_price": 3},
        {"name": "BOULT 25 X6", "quantity": height_divide_150 * 4, "unit_price": 0.04},
        {"name": "Self Lock Nut", "quantity": height_divide_150 * 8, "unit_price": 0.03},
        {"name": "Boult 20x6", "quantity": height_divide_150 * 4, "unit_price": 0.05},
        {"name": "SELF DRILL SCREW 7/7.16", "quantity": (scrap_add_50 / 300) * 4, "unit_price": 0.035},
        {"name": "Side Links", "quantity": height_divide_150 * 2, "unit_price": 6},
        {"name": "Spendal Rod 12.7 x65", "quantity": (height_divide_150 * 2) - 1, "unit_price": 1.5},
        {"name": "Rod 3000 x 12.7", "quantity": weight_50 * height_divide_150, "unit_price": 0},
        {"name": "Num Of Motor", "quantity": 1, "unit_price": 0},
        {"name": "beLIMO Motor", "quantity": number_of_motor * 0, "unit_price": 0},  
        {"name": "beLIMO Holder", "quantity": number_of_motor * 2, "unit_price": 0.5},  
        {"name": "Beliom Boult 75x6", "quantity": number_of_motor * 4, "unit_price": 40 / 1000},
        {"name": "Plain Nut", "quantity": number_of_motor * 4, "unit_price": 20 / 1000},
        {"name": "Spray Paint", "quantity": 2, "unit_price": 5},
    ]
    
    cost_breakdown = [
        {"component": comp["name"], "quantity": comp["quantity"], "unit_price": comp["unit_price"], "cost": comp["quantity"] * comp["unit_price"]}
        for comp in components
    ]
    total_cost = sum(item["cost"] for item in cost_breakdown)
    return total_cost, cost_breakdown

def calculate_lg1_cost(width, length):
    
    outer_frame_extrude_aluminum_quantity = (width + length) * 2 + 380
    extruded_aluminum_blade_quantity = round((width - 10) / 15) * length
    extruded_aluminum_pipe_quantity = round((length) / 350) * width
    powder_weight = 0.0000005
    powder_weight_input = length * width

    components = [
        {"name": "Outer Frame Extruded Aluminum", "quantity": outer_frame_extrude_aluminum_quantity / 1000, "unit_price": 3.3},
        {"name": "Extruded Aluminum Blade", "quantity": extruded_aluminum_blade_quantity / 1000, "unit_price": 2.0},
        {"name": "Extruded Aluminum Pipe 8 mm", "quantity": extruded_aluminum_pipe_quantity / 1000, "unit_price": 1.2},
        {"name": "Galvanized Corner", "quantity": 4, "unit_price": 0.5},
        {"name": "Accessories", "quantity": 1, "unit_price": 3},
        {"name": "Powder Weight", "quantity": powder_weight * powder_weight_input, "unit_price": 20},
    ]
    
    cost_breakdown = [
        {"component": comp["name"], "quantity": comp["quantity"], "unit_price": comp["unit_price"], "cost": comp["quantity"] * comp["unit_price"]}
        for comp in components
    ]
    total_cost = sum(item["cost"] for item in cost_breakdown)
    return total_cost, cost_breakdown

def calculate_lg1_selling_price(material_cost, width):
    # Constants for VCD-specific calculation
    participation_rate = 0.3
    
    # Powder time
    powder_time = (((6000 / width) * 4) * 3)
    powder_time_input = (((6000 / width) * 4) * 9)
    total_powder_time = ((90 / powder_time) + (120 / powder_time_input) + 1)
    powder_time_price = 2
    powder_time_cost = total_powder_time * powder_time_price
    
    direct_labor = (material_cost * participation_rate) + powder_time_cost
    overhead = direct_labor * 0.6 
    total_cost = material_cost + direct_labor + overhead

    # Profit margin
    profit_margin = 0.25
    selling_price = (total_cost * profit_margin) + total_cost

    product_code = "LG1-N-0"
    description = "Supply Linear Bar Grille, 0Â° Deflection Front Horizontal Fixed Bars without OBD"

    return {
        "total_cost": total_cost,
        "selling_price": selling_price,
        "product_code": product_code,
        "description": description,
    }

def calculate_cost_for_product(product_name, width, length, height, quantity):
    product_name = product_name.strip().lower()
    if product_name == "vcd":
        total_cost, components = calculate_vcd_cost(width, height)
        selling_details = calculate_vcd_selling_price(total_cost)
    elif product_name == "mfsd":
        total_cost, components = calculate_mfsd_cost(width, height)
        selling_details = calculate_mfsd_selling_price(total_cost)
    elif product_name == "fd":
        total_cost, components = calculate_fd_cost(width, height)
        selling_details = calculate_fd_selling_price(total_cost)
    elif product_name == "mvcd":
        total_cost, components = calculate_mvcd_cost(width, height)
        selling_details = calculate_mvcd_selling_price(total_cost)
    elif product_name == "lg1":
        total_cost, components = calculate_lg1_cost(width, length)
        selling_details = calculate_lg1_selling_price(total_cost, width)
    else:
        raise ValueError("Unknown product name.")

    # Scale material costs by quantity
    for component in components:
        component["quantity"] *= quantity
        component["cost"] = component["quantity"] * component["unit_price"]

    # Scale total cost and selling price by quantity
    total_cost *= quantity
    selling_details["total_cost"] *= quantity
    selling_details["selling_price"] *= quantity

    return total_cost, components, selling_details

def save_to_csv(products):
    filename = "material_cost.csv"
    with open(filename, mode="w", newline="") as file:
        writer = csv.writer(file)
        
        headers = ["Product", "Width", "Length", "Height", "Quantity", "Component", "Quantity Needed", "Unit Price", "Cost", "Total Cost"]
        
        for product in products:
            writer.writerow(headers)
            first_row = True
            for item in product["material_cost"]:
                writer.writerow([
                    product["name"] if first_row else "",
                    product["width"] if first_row and product["width"] > 0 else "",
                    product["length"] if first_row and product["length"] > 0 else "",
                    product["height"] if first_row and product["height"] > 0 else "",
                    product["quantity"] if first_row else "",
                    item["component"], item["quantity"], item["unit_price"], item["cost"], ""
                ])
                first_row = False
            
            writer.writerow(["", "", "", "", "", "", "", "", "", product["total_cost"]])
            writer.writerow([])

    print(f"Cost breakdown saved to {filename}")

def save_selling_prices_to_csv(products):
    filename = "selling_prices.csv"
    with open(filename, mode="w", newline="") as file:
        writer = csv.writer(file)
        
        # Headers
        headers = ["Product Code", "Description", "Width", "Length", "Height", "Quantity", "Final Selling Price"]
        writer.writerow(headers)
        
        # Write product details
        total_selling_price = 0  # Initialize total selling price
        for product in products:
            final_selling_price = product["selling_details"]["selling_price"]
            total_selling_price += final_selling_price  # Add to total
            
            writer.writerow([
                product["selling_details"]["product_code"],
                product["selling_details"]["description"],
                product["width"] if product["width"] > 0 else "",
                product["length"] if product["length"] > 0 else "",
                product["height"] if product["height"] > 0 else "",
                product["quantity"],
                round(final_selling_price, 2),
                ""  # Leave this blank for the individual rows
            ])
        
        # Add total selling price
        vat = round(total_selling_price * 0.15, 2)  # Calculate 15% VAT
        final_price_with_vat = round(total_selling_price + vat, 2)  # Final price after VAT
        
        # Add blank rows for better formatting
        writer.writerow([])  # Blank row for separation
        writer.writerow(["", "", "", "", "", "Total Selling Price", round(total_selling_price, 2)])
        writer.writerow(["", "", "", "", "", "VAT (15%)", vat])
        writer.writerow(["", "", "", "", "", "Final Price (With VAT)", final_price_with_vat])

    print(f"Selling prices with VAT saved to {filename}")

# Main logic
products = []
while True:
    product_name = input("Enter product name: ").strip()
    width = float(input("Enter width (0 if not applicable): "))
    length = float(input("Enter length (0 if not applicable): "))
    height = float(input("Enter height (0 if not applicable): "))
    quantity = int(input("Enter quantity: "))

    try:
        total_cost, components, selling_details = calculate_cost_for_product(product_name, width, length, height, quantity)
        products.append({
            "name": product_name,
            "width": width,
            "length": length,
            "height": height,
            "quantity": quantity,
            "material_cost": components,
            "total_cost": total_cost,
            "selling_details": selling_details,
        })
    except ValueError as e:
        print(e)

    more_products = input("Are there any more products? (yes/no): ").strip().lower()
    if more_products != "yes":
        break

save_to_csv(products)
save_selling_prices_to_csv(products)
```
